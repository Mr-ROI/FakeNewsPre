```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
import nltk
nltk.download('stopwords')
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
df.info() 
```

```python
df = df.fillna(' ')
```

```python
df['content'] = df['author'] + ' ' + df['title']
```

```python
df['content']
```

```python
def stemming(content):
    stem_port = PorterStemmer()
    
    stemmed_con = re.sub('[^a-zA-Z0-9]', ' ', content)
    stemmed_con = stemmed_con.lower()
    stemmed_con = stemmed_con.split()
    stemmed_content = [stem_port.stem(i) for i in stemmed_con if i not in stopwords.words('english')]
    stemmed_content = ' '.join(stemmed_content)
    return stemmed_content

```

```python
df['content'] = df['content'].apply(stemming)
```

```python
df['content']
```

```python
X = df['content'].values
y = df['label'].values
```

```python
vectorizer = TfidfVectorizer()

X = vectorizer.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y)
model = LogisticRegression()

model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_ = model.predict(X_test)

print('Accuracy score of the Model', accuracy_score(y_test, y_))
```

```python
import pickle 
with open('model.dat', 'wb') as file:
    pickle.dump(model, file)
```
